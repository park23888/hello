### Perceptrons - Making Predictions
#### Creating a gate with Perceptron

```python
import numpy as np #파이썬에 있는 것중 numpy라는 것을 꺼낸다.
```

### AND Gate

```python
def AND(x1, x2): #and게이트는 둘다 넣는다.
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.7
    #activation계산 
    tmp = w[0]*x[0] + w[1]*x[1] + b # tmp = np.sum(w*x) + b
    print(tmp)
    if tmp <= 0: 
      return 0
    else: #else=every thing
        return 1
```

```python
AND(1,0)
```

```python
AND(1,1)
```

```python
AND(0,0)
```

```python
AND(0,1)
```

### NAND Gate

```python
def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5])
    b = 0.7
    tmp = w[0]*x[0] + w[1]*x[1] + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
NAND(0,0)
```

```python
NAND(1,0)
```

```python
NAND(1,1)
```

### OR Gate

```python
def OR(x1, x2): #b만 바꿨는데 동작이 왜되는가?? 시험에 나올 확률 높음
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.2
    tmp = np.sum(w*x) + b
    print(tmp)
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
OR(0,0)
```

```python
OR(1,0)
```

```python
OR(0,1)
```

```python
OR(1,1)
```

### XOR Gate

```python
def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    return y
```

```python
XOR(0,0)
```

```python
XOR(1,1)
```

```python
XOR(0,1)
```

```python
XOR(1,0)
```

XOR cannot be expressed as a single layer Perceptron.
